```python
from blackjack_module import MonteCarloBlackjack
import warnings
```

```python
print(MonteCarloBlackjack.__doc__)
```

```python
print("Running On-Policy MC Control...")
bj_onpolicy = MonteCarloBlackjack(num_episodes=500000, epsilon=0.1, alpha=1/500, seed=0)
bj_onpolicy.train()
```

```python
print("Plotting Action-Value Heatmap...")
bj_onpolicy.plot('Q', width=310, height=250)
```

```python
print("Plotting State-Action Counts...")
bj_onpolicy.plot('C', width=325, height=250)
```

```python
print("Plotting Hit/Stick Values Over Episodes...")
df = bj_onpolicy.history.query(
    "(player_sum==17) & (dealer_card==9) & (usable_ace==False)"
)
df_chart = df[['Q','action','episode']]
```

```python
import altair as alt
alt.Chart(df_chart).mark_line().encode(
    x='episode', y='Q', color='action'
).properties(width=400, height=300)
```
